```python
df=df.sort_values(by='Payment_Method')
df
```

```python
df['Payment_Method']=df['Payment_Method'].fillna(df['Payment_Method'].mode()[0])      #missing value with mode
```

```python
df.rename(columns={'Order_ID':'OrderID'})
```

```python
df['Date'] = pd.to_datetime(df['Date'])     # convert date column to datetime
```

```python
df['Month'] = df['Date'].dt.month_name()      # create a month column
```

# Univariate analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

### Distribution of total amount

```python
sns.distplot(df['Total_Amount'])
plt.show()
```

### Insights
The distplot shows that the Total Amount is right-skewed, meaning most customers spend lower amounts, while only a few customers make very
high purchases. This indicates that high-value transactions areless frequent but significantly impact overall revenue. The distribution 
highlights the importance of targeting both regular buyers and high-value customers to maximize sales.

### To find the averge age of customers

```python
sns.boxplot(df['Age']);
```

### Insights
The box plot indicates the central age of customers and highlights any outliers. Most customers fall within the interquartile range,
while any points outside this range represent unusually young or old customers compared to the rest of the data.


### Customer Rating Frequency distribution

```python
sns.histplot(df['Customer_Rating']);
```

### Insights
The histogram shows that most customers give higher ratings, indicating generally positive satisfaction with the products or services.
Lower ratings occur less frequently, meaning only a small portion of customers report negative experiences. Overall, the distribution
suggests consistently good customer feedback.

###  Most frequently used payment Method

```python
order=df['Payment_Method'].value_counts().index
sns.countplot(x=df['Payment_Method'],order=order)     #gives payment methods sorted by their frequency
plt.xticks(rotation=45)
plt.show()
```

### Insights
The count plot shows that Most Used Payment Method is the most preferred by customers, indicating a clear preference for this mode of payment. In contrast, Least Used Payment Method is rarely chosen, suggesting it is less convenient or trusted. Overall, the distribution highlights that a few payment methods dominate customer choices, which can help optimize payment options and marketing strategies.

```python
chan_sales = df.groupby('Product_Category')['Total_Amount'].sum().sort_values(ascending=False)
plt.figure(figsize=(5, 5))

# Plot pie chart with percentage labels and a defined start angle
plt.pie(
    chan_sales.values,                   
    labels=chan_sales.index,            
           
    startangle=140,                     
    colors=sns.color_palette('coolwarm') 
)

plt.title('Total Sales by product')
plt.tight_layout()
plt.show()
```

##  Monthly Sales Trend Over Time


```python
# Convert order_date to monthly period (e.g., 2022-01, 2022-02)
df['order_month'] = df['Date'].dt.to_period('M')
```

```python
monthly_sales=df.groupby('order_month')['Total_Amount'].sum()
```

```python
plt.figure(figsize=(15,4))
monthly_sales.plot(marker='o', color='navy')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total sales')
plt.xticks(rotation=45)
# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

### Insights
 The monthly sales trend shows clear variations across different months.
Certain months experience higher sales, indicating possible seasonal demand or promotional influence.
This trend helps businesses plan inventory and marketing strategies more effectively.



## Sales by Product Category

```python
plt.figure(figsize=(10,5))
order = df.groupby('Product_Category')['Total_Amount'].sum().sort_values(ascending=False).index
sns.barplot(data=df,x='Product_Category',y='Total_Amount',estimator=sum,order=order)
plt.title('Sales by Product Category')
plt.xticks(rotation=45)
plt.show()
```

### Insights
From the visual, it is clear that Electronics is the highest-selling product category by a large margin, contributing significantly more 
revenue than all other categories. This indicates a very strong customer demand for electronic items and suggests that this category plays
a major role in overall business performance. Home & Garden and Sports form the second tier of sales, showing moderate performance and 
consistent interest from customers. In contrast, categories such as Fashion, Toys, Beauty, Food, and Books generate comparatively lower sales, 
with only small differences among them. This uneven distribution highlights that revenue is concentrated in only a few key categories.
To improve overall performance, the business should continue strengthening its top-performing categories while also exploring better
marketing strategies, promotions, or product improvements to boost the lower-performing ones.

## Average Customer Rating by City

```python
city_rating = df.groupby('City')['Customer_Rating'].mean().sort_values(ascending=False).reset_index()

plt.figure(figsize=(10,5))
sns.barplot(data=city_rating, x='City', y='Customer_Rating',palette='viridis')
plt.title('Average Rating by City')
plt.xticks(rotation=45)
plt.show()
```

### Insights
The average customer rating shows noticeable variation across different cities. Some cities consistently report higher ratings,
indicating better customer satisfaction, service quality, or delivery experience in those areas. In contrast, a few cities display
relatively lower average ratings, which may suggest issues such as delayed deliveries, product dissatisfaction, or gaps in customer
support. Overall, this pattern highlights the need for location-specific improvements, where cities with lower ratings should be 
prioritized for service enhancement while best-performing cities can be used as benchmarks for maintaining high standards.

## Orders by Device Type

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Device_Type',palette='magma')
plt.title('Orders by Device Type')
plt.show()
```

### Insights
Orders placed through mobile devices dominate compared to desktop and tablet, showing that most customers prefer the convenience 
of shopping on their phones. This indicates a strong shift toward mobile-first behavior, likely due to ease of access, app usage,
and quick checkout options. Desktop users still contribute a moderate share of orders, suggesting it remains relevant for more detailed
browsing or bulk purchases. Tablets account for the least number of orders, implying lower preference or limited usage for online shopping. 
Overall, the results highlight the importance of optimizing the shopping experience for mobile users, as they represent the primary source
of orders.

### Key Insights

1. Female customers contribute a higher portion of total sales, indicating strong purchasing power and higher engagement compared to other genders.


2. Mobile devices are the most commonly used platform for purchases, showing that the business should prioritize mobile optimization and marketing.


3. Certain product categories perform significantly better than others, suggesting customer preference is concentrated in a few high-demand categories.


4. Some cities generate higher revenue but also show higher return rates, which may indicate issues in product suitability or delivery experience in those regions.


5. Longer session duration is generally associated with higher customer ratings, suggesting that better user experience and engagement lead to greater satisfaction.


6. A few payment methods dominate transactions, showing customer trust in specific payment channels and the opportunity to streamline others.


7. Clear seasonal and monthly trends in sales indicate peak periods that can be targeted for promotions and stock planning.


8. Lower-performing customer segments and locations represent potential growth opportunities if supported with better targeting, offers, and service improvements.

## Recommendations

1. The company should increase focus on high-performing customer segments (such as the age group, gender, or city contributing the most revenue) by offering targeted promotions and personalized discounts.


2. Product categories that show consistently high sales should be prioritized in inventory planning to avoid stock-outs and maximize revenue potential.


3. For categories or cities with low sales or high return rates, the company should investigate possible issues such as product quality, delivery delays, or mismatched customer expectations.


4. Marketing efforts should be optimized by focusing more on the most commonly used device type and payment method to improve customer convenience and increase conversion rates.


5. Improving user experience for sessions with shorter duration might help boost customer ratings and overall satisfaction.


6. Regular monitoring of customer feedback and return patterns is recommended to quickly identify and solve recurring problems.


7. Seasonal trends observed in the data should be used for better planning of sales campaigns and promotional offers during peak periods.
